# EXPLORATORY DATA ANALYSIS (EDA)

Exploratory data analysis (EDA) is a very important step because it is very important to be able to understand the nature of the data without making assumptions. The results of data exploration can be extremely useful in grasping the structure of the data, the distribution of the values, and the presence of extreme values and interrelationships within the data set. The purpose of EDA is to use summary statistics and visualizations to better understand data, find clues about the tendencies of the data.

**Loading the libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
breast_cancer = pd.read_csv(r"C:\Users\maria\Desktop\proyecto cancer de mama\breast-cancer-wisconsin-data_data.csv")
breast_cancer.head()
```

**Data visualization using Boxplots**

**Heatmap**

Let's create a heatmap to visualize if there is correlation among the different features (excluding the diagnosis, as it is not continuous).

```python
plt.figure(figsize=(20,12)) 
sns.heatmap(breast_cancer.drop('diagnosis', axis=1).corr(), annot=True) 
```

Obviously, the diagonal has a correlation coefficient of 1 because it represents the correlation between a feature and itself. We observe a direct correlation among several parameters, such as radius_mean, radius_worst, radius_se, area_worst, area_mean, perimeter_mean, perimeter_se, and perimeter_worst. Let's create a plot to represent the correlation among those variables.

```python
cols=['radius_mean', 'radius_worst', 'radius_se', 'area_worst', 'area_mean', 'perimeter_mean', 'perimeter_se', 'perimeter_worst']
sns.pairplot(breast_cancer[cols])
```
